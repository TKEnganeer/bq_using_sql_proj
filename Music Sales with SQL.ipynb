## Answering Business Questions for a Music Sales Company with SQL

```python
# importing the relevant python modules
import sqlite3
import pandas as pd   # allows the writing of SQL queries to produce dataframe outputs
import numpy as np
import matplotlib.pyplot as plt # graphicla outputs
from matplotlib import cm
%matplotlib inline
```

```python
'''Helper Functions'''  

db = 'chinook.db'

def run_query(q):
    with sqlite3.connect(db) as conn:
        return pd.read_sql(q, conn)          # connection object created and read 
    
def run_command(command):
    with sqlite3.connect(db) as conn:
        conn.isolation_level = None
        conn.execute(command)                # SQL queries can now be run

def show_tables():
    q ='''
    SELECT 
        name,
        type
    FROM sqlite_master
    WHERE type IN ("table", "view")
    '''
    return run_query(q)                    


show_tables()                              # state of the database
```

## 1. Tracks in the USA Based on Genre

In this section I calculated the most popular tracks in the USA by genre based on the amount of tracks sold. I also determined their popularity as a percentage of total sales. The goal was to determine which other artists tracks worth purchasing for sales based on their genre. The choice of artists is shown in the table below.

|Artist Name | Genre |
|------------|-------|
|Regal |	Hip-Hop |
|Red Tone	|Punk|
|Meteor and the Girls|	Pop|
|Slim Jim Bites |	Blues|


```python
albums_to_purchase ='''
WITH track_genre_USA AS
    (
        SELECT
            i.billing_country country,
            t.genre_id,
            g.name music_genre,
            COUNT(il.quantity) tracks_sold    
        FROM genre g
        INNER JOIN invoice i ON i.invoice_id = il.invoice_id
        INNER JOIN invoice_line il ON il.track_id = t.track_id 
        INNER JOIN track t ON t.genre_id = g.genre_id
        WHERE country = "USA"
        GROUP BY 3
    )

SELECT
    genre_id,
    music_genre,
    tracks_sold no_tracks_sold,
    CAST(tracks_sold AS FLOAT)/
    CAST((SELECT SUM(tracks_sold) FROM track_genre_USA) AS FLOAT)*100 AS percentage_sold
FROM track_genre_USA
GROUP BY 2
ORDER BY 3 DESC
LIMIT 10;
'''
run_query(albums_to_purchase)                    # part 1a - 

```

## 1. Analysis 

The table shows that Rock music sells the most in the US with over 50% of sales. Punk, Blues and Pop account for 12.3%, 3.4% and 2.1% respectively. Therefore, I would recommend that the 3 artist worth purchasing would be Red Tone, Meteor and the Girls and Slim Kim Bites based on this data. 

```python
genre_sales_usa = run_query(albums_to_purchase)
genre_sales_usa.set_index("music_genre", inplace=True, drop=True)

genre_sales_usa["no_tracks_sold"].plot.barh(
    title="Top 10 Selling Genres in the USA",
    xlim=(0, 625),
    colormap=plt.cm.Accent
)
plt.ylabel('Genre')

for i, label in enumerate(list(genre_sales_usa.index)):
    score = genre_sales_usa.loc[label, "no_tracks_sold"]
    label = (genre_sales_usa.loc[label, "percentage_sold"]
            ).astype(int).astype(str) + "%"
    plt.annotate(str(label), (score + 10, i - 0.15))

plt.show()
```

In this section I determined the best performing sales agent based on their support_rep_id and total sales. I also used the date that they were hired by the company to get a better understanding of the results.

```python
top_sales_agent = '''
WITH customer_rep_sales AS
     (
      SELECT 
            c.support_rep_id support_agent,
            SUM(i.total) total_sales
       FROM invoice i
       INNER JOIN customer c ON c.customer_id = i.customer_id
       GROUP BY 1
     ) 


SELECT 
        e.first_name || " " || e.last_name agent_name,
        e.hire_date,
        crs.total_sales rep_dollar_sales
   FROM customer_rep_sales crs
  INNER JOIN employee e ON e.employee_id = crs.support_agent
  GROUP BY 1
  ORDER BY 3;
'''
    
run_query(top_sales_agent)  
```

## 3. Sales Data in Worldwide

In this section I analyzed the sales data for all the companies customers around the world. The goal was to determine which countries as a market had the best potential for future growth based on the following results:
- total number of customers
- total value of sales
- average value of sales per customer
- average order value

```python
sales_per_country = '''
WITH country_other AS 
    (
     SELECT
       CASE
           WHEN (
                 SELECT count(*)
                 FROM customer 
                 where country = c.country
                ) = 1 THEN "Other"
           ELSE c.country
       END AS country,
       c.customer_id,
       il.*
     FROM invoice_line il
     INNER JOIN invoice i ON i.invoice_id = il.invoice_id
     INNER JOIN customer c ON c.customer_id = i.customer_id
    )
    
SELECT
    country,
    customers,
    total_sales,
    average_order_val,
    val_sales_per_customer
FROM 
    (
    SELECT
        country,
        count(DISTINCT customer_id) customers,
        SUM(unit_price) total_sales,
        SUM(unit_price)/count(DISTINCT invoice_id) average_order_val,
        SUM(unit_price)/count(DISTINCT customer_id) val_sales_per_customer,
        CASE
            WHEN country = "Other" THEN 1
            ELSE 0
        END AS sort
    FROM country_other
    GROUP BY country
    ORDER BY sort ASC, total_sales DESC
    );
'''

run_query(sales_per_country)
```

```python
customer_purchases = run_query(sales_per_country)
customer_purchases.set_index("country", inplace=True, drop=True)
colors = [plt.cm.Accent(i) for i in np.linspace(0, 1, customer_purchases.shape[0])]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9, 10))
ax1, ax2, ax3, ax4 = axes.flatten()
fig.subplots_adjust(hspace=.7, wspace=.4)

# top left
customer_purchases["customers"].plot.bar(
    ax=ax1,
    title="Customers Sales Worldwide",
    xlim=(0, 20),
    color=colors
)
# plt.ylabel('Country')

for i, label in enumerate(list(customer_purchases.index)):
    score = customer_purchases.loc[label, "customers"]
#     label = (customer_purchases.loc[label, "percentage_sold"]
#             ).astype(int).astype(str) + "%"
#     plt.annotate(str(label), (score + 10, i - 0.15))

# top right
sales_distribution = customer_purchases["total_sales"].copy().rename('')
sales_distribution.plot.pie(
    ax=ax2,
    startangle=-90,
    counterclock=False,
    title='Customer Sales Distribution',
    colormap=plt.cm.Accent,
    fontsize=8,
    wedgeprops={'linewidth':0}
    
)
# bottom left
avg_order = customer_purchases["average_order_val"].copy()
avg_order.index.name = ''
difference_from_avg = avg_order * 100 / avg_order.mean() - 100
difference_from_avg.drop("Other", inplace=True)
difference_from_avg.plot.bar(
    ax=ax3,
    color=colors,
    title="Average Order,\n % Difference from Mean"
)
ax3.tick_params(top="off", right="off", left="off", bottom="off")
ax3.axhline(0, color='k')
ax3.spines["top"].set_visible(False)
ax3.spines["right"].set_visible(False)
ax3.spines["bottom"].set_visible(False)

# bottom right
cv = customer_purchases["val_sales_per_customer"].copy()
cv.index.name = ''
cv.drop("Other",inplace=True) # removed conuntries with only 1 customer
cv.plot.bar(
    ax=ax4,
    color=colors,
    title="Customer Value, Dollars"
)
ax4.tick_params(top="off", right="off", left="off", bottom="off")
ax4.spines["top"].set_visible(False)
ax4.spines["right"].set_visible(False)

plt.show()
```

## 4. Tracks vs Album Purchases 

In this section I compared album sales against tracks (from the same album) sales in the database based in the invoice data. The goal was to understand whether it is more financially advantageous buy whole albums from the record labels or for the company to sell or buy the popular tracks from the albums indivually. So I wanted to know how many tracks were purchased compared to whole albums in terms of real numbers and percentages. 

```python
track_albums = '''

WITH
    invoice_first_track AS
    (
    SELECT
        il.invoice_id,
        MIN(il.track_id) first_track_id
        FROM invoice_line il
        GROUP BY 1
    )

SELECT
    album_purchase,
    COUNT(invoice_id) numb_invoices,
    CAST (COUNT(invoice_id) AS FLOAT)/(SELECT COUNT(*) FROM invoice)*100 
    AS percentage

FROM
    (
    SELECT
        ifs.*,
        CASE
            WHEN
                (
                SELECT t.track_id FROM track t
                WHERE t.album_id = (
                                    SELECT t2.album_id FROM track t2
                                    WHERE t2.track_id = ifs.first_track_id
                                    )
                EXCEPT
                
                SELECT il2.track_id FROM invoice_line il2 
                WHERE il2.invoice_id = ifs.invoice_id)    
                
                IS NULL 
                
                AND 
                
                (
                SELECT il2.track_id FROM invoice_line il2 
                WHERE il2.invoice_id = ifs.invoice_id
                
                EXCEPT
                
                SELECT t.track_id FROM track t
                WHERE t.album_id = (
                                    SELECT t2.album_id FROM track t2
                                    WHERE t2.track_id = ifs.first_track_id
                                    )) 
                IS NULL
                
                                    
            THEN "yes"
            ELSE "no"
         END AS "album_purchase"
        FROM invoice_first_track ifs
        )
GROUP BY album_purchase;

'''

run_query(track_albums)
```

## 4. Analysis 

From the table it is clear that the majority of the company's revenue come from tracks and not whole album purchases. This could mean that the company might be better of buying specific tracks which are most popular from the record label as oppossed to the whole album. On the other hand if the company could loose potentially 20% of its revenue if they were to exlucde whole album purchases. This has been further illustrated in the pie chart below.
